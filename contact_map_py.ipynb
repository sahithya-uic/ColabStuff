```python
import matplotlib.pyplot as plt
```

```python
import MDAnalysis as mda
```

```python
u=mda.Universe('29_protein.pdb')
```

```python
chainA = u.select_atoms("segid A and name CA")
chainB = u.select_atoms("segid B and name CA")
```

```python
len(chainA)
```

```python
len(chainB)
```

```python
from MDAnalysis.core.groups import distances
dist_arr = distances.distance_array(chainA.positions,chainB.positions)
dist_arr.shape
```

```python
fig,ax=plt.subplots()
im = ax.imshow(dist_arr,origin='lower')
cbar = fig.colorbar(im)
cbar.ax.set_ylabel('Distance (Angstrom)')
plt.show()
```
